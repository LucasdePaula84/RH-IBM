# Carregando bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

# Carregando Dataset

```python
df = pd.read_csv ('Documents/Python/WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

# Informações basicas

```python
df.info()
```

# Analise Univariada

```python
_ = plt.hist(df.Age)
_ = plt.title('Distribuicao de Idade')
```

```python
_ = plt.hist(df.MonthlyIncome)
_ = plt.title('Distribuicao Salario')
```

```python
df.Gender.value_counts(1)
```

```python
df.EducationField.value_counts(1)
```

```python
_ = df.EducationField.value_counts(1).plot.bar()
_ = plt.title('Head Count por Campo Educacional')
```

# Analise Bivariada

```python
_ = sns.boxplot(x= 'Gender', y= 'MonthlyIncome', data=df)
_ = plt.title('Distribuicao Salarial por Sexo')
```

# Analise Multivariada

```python
_ = sns.scatterplot(x= 'YearsAtCompany', y= 'MonthlyIncome', data=df)
_ = plt.title('Distribuicao Salarial por Tempo de Casa')
```
